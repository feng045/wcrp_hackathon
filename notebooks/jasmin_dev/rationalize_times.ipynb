```python
def find_main_time(ds):
    times = {c: pd.DatetimeIndex(ds2d[c].values)
             for c in ds2d.coords if c.startswith('time')}
    for n, time in times.items():
        if ((time.second == 0) & (time.minute == 0)).all():
            return n

def find_halfhour_time(ds):
    times = {c: pd.DatetimeIndex(ds2d[c].values)
             for c in ds2d.coords if c.startswith('time')}
    for n, time in times.items():
        if ((time.second == 0) & (time.minute == 30)).all():
            return n
```

```python
ds2d_new = xr.Dataset()
for name, da in ds2d.data_vars.items():
    if 'time' in da.coords or 'time_1' in da.coords:
        ds2d_new[name] = da  
```

```python
def align_to_main_time(time, da, timename):
    reduced_dims = [d for d in da.dims if d != timename]
    reduced_coords = {d: da[d] for d in reduced_dims}
    dims = ['time'] + reduced_dims
    coords = {'time': time, **reduced_coords}

    if len(time) > len(da[timename]):
        init_idx = np.argmin(np.abs((time.values - ds2d[timename].values[0])))
        assert np.abs((time.values[slice(init_idx, None)] - ds2d[timename].values) / 1e9).max() < 300

        dim_shape = [v for k, v in da.sizes.items() if k != timename]

        new_data = np.zeros((len(time), *dim_shape))
        new_data[slice(init_idx, None)] = da.values
        new_da = xr.DataArray(new_data, dims=dims, coords=coords)
    elif len(time) == len(da[timename]):
        #reduced_da = da.drop_vars(timename)
        assert np.abs((time.values - ds2d[timename].values) / 1e9).max() < 300
        new_da = xr.DataArray(da.values, dims=dims, coords=coords)
    return new_da
```
