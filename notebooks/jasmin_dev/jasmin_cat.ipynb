```python
import intake
cat = intake.open_catalog('https://raw.githubusercontent.com/digital-earths-global-hackathon/catalog/refs/heads/main/UK/main.yaml')
```

```python
list(cat)
```

```python
sim = cat['um_glm_n2560_RAL3p3']
```

```python
ds = sim(zoom=9).to_dask()
```

```python
{n: v.attrs['long_name'] for n, v in ds.data_vars.items()}
```

```python
def long_names(ds):
    for n, da in ds.data_vars.items():
        da.attrs['short_name'] = da.name
    return ds.rename({n: da.attrs['long_name'] for n, da in ds.data_vars.items()})
ds.pipe(long_names)
```
