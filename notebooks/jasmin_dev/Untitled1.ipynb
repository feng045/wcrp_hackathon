```python
import cartopy.crs as ccrs
import cartopy.feature as cf
import matplotlib.pyplot as plt
import xarray as xr

import easygems.healpix as egh
```

```python
ds2 = xr.open_dataset('http://hackathon-o.s3.jc.rl.ac.uk/sim-data/DYAMOND3_example_data/sample_data_hirerarchy/5km-RAL3/2d.z10.zarr', engine='zarr')
```

```python
ds2
```

```python
projection = ccrs.PlateCarree()
figsize = (8, 4)

fig, ax = plt.subplots(
    figsize=figsize, subplot_kw={"projection": projection}, constrained_layout=True
)
ax.set_global()
mean = ds2.rlut.isel(time=0).mean()
ax.set_title(f'OLR global mean = {mean:.8f} W/m2')

egh.healpix_show(ds2.rlut.isel(time=0), ax=ax)
ax.add_feature(cf.COASTLINE, linewidth=0.8)
ax.add_feature(cf.BORDERS, linewidth=0.4)
```
