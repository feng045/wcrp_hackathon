```python
def has_dimensions(*dims):
    """Returns an Iris constraint that filters cubes based on dimensions."""
    def dim_filter(cube):
        cube_dims = [c.name() for c in cube.dim_coords]
        return set(cube_dims) == set(dims)
    return iris.Constraint(cube_func=dim_filter)
```

```python
# Example usage
constraint_2d = has_dimensions("time", "latitude", "longitude")
constraint_3d_pressure = has_dimensions("time", "pressure", "latitude", "longitude")
constraint_3d_model_level = has_dimensions("time", "model_level_number", "latitude", "longitude")
```

```python
def interp_to_standard_grid(ds, varname):
    lat = ds.latitude
    lon = ds.longitude
    da = ds[varname]
    time_coord = ds[[c for c in da.coords if c.startswith('time')][0]]
    lat_coord_name = [c for c in da.coords if c.startswith('latitude')][0]
    lon_coord_name = [c for c in da.coords if c.startswith('longitude')][0]
    
    data = np.zeros([*da.shape[:-2]] + [len(lat), len(lon)])
    new_da = xr.DataArray(
        data,
        name=da.name,
        coords=dict(
            time_0=time_coord,
            latitude=lat,
            longitude=lon,
        )
    )
    new_da.values = da.interp(**{lat_coord_name:lat, lon_coord_name:lon}).values
    ds[varname] = new_da
    
```
